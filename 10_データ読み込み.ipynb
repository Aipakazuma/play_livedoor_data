```python
import xml.etree.ElementTree as et
import pandas as pd

class XML2DataFrame:

    def __init__(self, xml_data):
        self.root = et.XML(xml_data)

    def parse(self, root):
        parses = []
        docs = list([child for child in iter(self.root)])
        for doc in docs:
            children = list(doc)
            parsed = dict()
            for child in children:
                name = child.attrib.get('name')
                if not name in list(parsed.keys()): parsed[name] = ''
                parsed[name] += '' if child.text is None else child.text

            parses.append(parsed)
        
        return parses

    def process_data(self):
        structure_data = self.parse(self.root)
        return pd.DataFrame(structure_data)


def xml2df(file_path):
    with open(file_path, 'r') as f:
        xml_data = f.read()

    xml2df = XML2DataFrame(xml_data)
    df = xml2df.process_data()
    return df
```
